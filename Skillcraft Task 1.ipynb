Aim

Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, 
such as the distribution of ages or genders in a population. 


```python
# Import libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading dataset
df = pd.read_csv('income.csv')
```

```python
# Handling Null Values

df['occupation'] = df['occupation'].replace(' ?' , 'Unknown')

print(df.head())
```

```python
# Understand data and its type
print(df.info())
print(df.describe())
```

```python
plt.bar(df['occupation'] , df['age'])
plt.title('JobType Distribution')
plt.xlabel('JobType')
plt.ylabel('Age')
plt.xticks(rotation=90)
plt.ylim(40,100)

plt.figure(figsize=(200,100))

plt.show()
plt.savefig('bargraph.png')
```

```python
plt.hist(df['age'],bins=20, color='orange', edgecolor='white')
plt.title('Age Distribution')

plt.xlabel('JobType')
plt.ylabel('Age')
plt.show()
```

Summary:

 **Bar Graph**:
The bar graph highlights that young individuals are predominantly employed in professions such as the armed forces 
and technical support. These fields appear to attract a younger demographic, indicating a trend where physically 
demanding or tech-oriented roles are more popular among younger people.

 **Histogram**:
The histogram reveals that younger age groups tend to have higher income levels compared to older age groups. 
This suggests that younger individuals, likely in their peak earning years or working in high-demand fields, 
are earning more than their older counterparts, who might have retired or transitioned to less lucrative roles.


```python
# Distribution of ages in a population.

df = pd.read_csv('Age_Group_Population.csv')

plt.bar(df['Age Group'], df['Population (Mn)'], color=['yellow', 'blue', 'pink'], edgecolor='black')

plt.title("India's Population Distribution by Age in 2022")
plt.xlabel("Age Group")
plt.ylabel("Population (in Millions)")

for i, value in enumerate(df['Population (Mn)']):
    plt.text(i, value + 10, f"{df['Percentage'][i]}%", ha='center', fontsize=10)

plt.show()
```
